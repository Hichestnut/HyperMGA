```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data_iga = np.loadtxt('iga_cm_cora.txt')
data_fga = np.loadtxt('fga_cm.txt')
data_ours = np.loadtxt('ours_cm_cora.txt')
data_dice = np.loadtxt('dice_cm.txt')
data_delete = np.loadtxt('delete_cm.txt')
data_random = np.loadtxt('random_cm.txt')
data = [data_random, data_delete, data_dice, data_fga, data_iga, data_ours]
data = np.array(data)
data = data.T
data2 = pd.DataFrame(data, columns=['Random','Delete','Dice','Fga','Iga','Ours'])
print(data)
print(data.shape)
data2.columns
```

```python
print(plt.rcParams["font.family"])
print(plt.rcParams['font.sans-serif'])
```

```python
ax = plt.figure()  # 新建画图板
sns.set_style('white')  # 设置风格

# rc = {'font.sans-serif': ['Arial']}
sns.boxplot(data=data2) # 绘制盒图
sns.set(font='Arial',font_scale =0.4)
# ax.set_xlabel(fontsize=14)
# plt.title('cora')  # 设置标题
plt.xlabel("attack")
plt.savefig('1.png')
```
